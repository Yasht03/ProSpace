```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('/kaggle/input/movielens-full-25-million-recommendation-data/ml-25m/movies.csv')
```

```python
movies.head()
```

```python
tfidf = TfidfVectorizer()
v = tfidf.fit_transform(movies['genres'])
all_features = tfidf.get_feature_names_out()
vocab = tfidf.vocabulary_
b = tfidf.fit_transform(movies['genres']).toarray()
```

```python
vocab
```

```python
v.shape
```

```python
b[0]
```

```python
cos = cosine_similarity(b)
```

```python
print(cos.shape)
```

```python
sorted(list(enumerate(cos[0])),reverse=True,key = lambda x:x[1] )
```

```python
def recommend(mov):
    movie_index = movies[movies['title']== mov].index[0]
    distances = cos[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1] )[1:11]
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
    
```

```python
recommend('Jumanji (1995)')
```
